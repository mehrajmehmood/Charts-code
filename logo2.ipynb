```python
l = ['amina', 'amina', 'hajra', 'safia', 'noman.', 'ali', 'ali', 'ali', 'ali', 'ali', 'ali', 'ali', 'naman', 'nimra', "ayesha"]

# 1. Type of the variable
list_type = type(l)
print(f"Type of the variable: {list_type}")

# 2. Number of elements
num_elements = len(l)
print(f"Number of elements: {num_elements}")

# 3. Unique values
unique_values = set(l)
print(f"Unique values: {unique_values}")

# 4. Convert to tuple
unique_tuple = tuple(unique_values)
print(f"Tuple of unique values: {unique_tuple}")
```
